```python
import random, shutil
from pathlib import Path

SEED = 42
TRAIN_RATIO, VAL_RATIO, TEST_RATIO = 0.70, 0.15, 0.15
assert abs(TRAIN_RATIO + VAL_RATIO + TEST_RATIO - 1.0) < 1e-9

OUT_ROOT = Path("/content/dataset")
if OUT_ROOT.exists():
    shutil.rmtree(OUT_ROOT)
for split in ["train","val","test"]:
    (OUT_ROOT/split/"correct").mkdir(parents=True, exist_ok=True)
    (OUT_ROOT/split/"wrong").mkdir(parents=True, exist_ok=True)

def split_list(items):
    items = list(items)
    random.shuffle(items)
    n = len(items)
    n_train = int(n * TRAIN_RATIO)
    n_val   = int(n * VAL_RATIO)
    train = items[:n_train]
    val   = items[n_train:n_train+n_val]
    test  = items[n_train+n_val:]
    return train, val, test

def move_files(files, dest_dir):
    dest_dir = Path(dest_dir)
    for f in files:
        dst = dest_dir / f.name
        # Avoid collisions
        if dst.exists():
            stem, suf = f.stem, f.suffix
            i = 1
            while (dest_dir / f"{stem}_{i}{suf}").exists():
                i += 1
            dst = dest_dir / f"{stem}_{i}{suf}"
        shutil.move(str(f), str(dst))

def collect_videos(folder: Path):
    # collect common video extensions
    exts = ["*.mp4","*.avi","*.mov","*.mkv","*.webm"]
    vids = []
    for e in exts:
        vids.extend(folder.glob(e))
    return sorted(vids)

random.seed(SEED)

correct_videos = collect_videos(SRC_CORRECT)
wrong_videos   = collect_videos(SRC_WRONG)

print("Correct videos:", len(correct_videos))
print("Wrong videos:  ", len(wrong_videos))

c_tr, c_va, c_te = split_list(correct_videos)
w_tr, w_va, w_te = split_list(wrong_videos)

print("\nSplit counts")
print("Correct:", len(c_tr), len(c_va), len(c_te))
print("Wrong:  ", len(w_tr), len(w_va), len(w_te))

move_files(c_tr, OUT_ROOT/"train"/"correct")
move_files(c_va, OUT_ROOT/"val"/"correct")
move_files(c_te, OUT_ROOT/"test"/"correct")

move_files(w_tr, OUT_ROOT/"train"/"wrong")
move_files(w_va, OUT_ROOT/"val"/"wrong")
move_files(w_te, OUT_ROOT/"test"/"wrong")

print("\n✅ Done. Example listing:")
!find /content/dataset/train -maxdepth 2 -type f | head -n 10
```

```python
import cv2
import numpy as np

def sample_frames_uniform(video_path: str, num_frames: int = 16):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        raise RuntimeError(f"Cannot open video: {video_path}")

    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if total <= 0:
        # fallback: read sequentially
        frames = []
        while True:
            ok, frame = cap.read()
            if not ok:
                break
            frames.append(frame)
        cap.release()
        if len(frames) == 0:
            raise RuntimeError(f"No frames in video: {video_path}")
        total = len(frames)
        idxs = np.linspace(0, total-1, num_frames).astype(int)
        return [frames[i] for i in idxs]

    idxs = np.linspace(0, max(total-1,0), num_frames).astype(int)
    out = []
    for i in idxs:
        cap.set(cv2.CAP_PROP_POS_FRAMES, int(i))
        ok, frame = cap.read()
        if not ok:
            # fallback: just break and later pad
            break
        out.append(frame)
    cap.release()

    # pad if short
    if len(out) == 0:
        raise RuntimeError(f"Failed to read frames: {video_path}")
    while len(out) < num_frames:
        out.append(out[-1].copy())
    return out

# Quick sanity check on one video
example = next((OUT_ROOT/"train"/"correct").glob("*.mp4"))
frames = sample_frames_uniform(str(example), num_frames=16)
print("Sampled frames:", len(frames), "shape:", frames[0].shape)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms

IMG_SIZE = 224
T = 16  # frames per video

# ImageNet normalization
img_tf = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]),
])

class VideoFolderDataset(Dataset):
    def __init__(self, root_dir: str, split: str):
        self.root = Path(root_dir) / split
        self.samples = []
        # label map
        classes = [("correct", 0), ("wrong", 1)]
        for cname, label in classes:
            for vid in (self.root / cname).glob("*"):
                if vid.suffix.lower() in [".mp4",".avi",".mov",".mkv",".webm"]:
                    self.samples.append((vid, label))

        if len(self.samples) == 0:
            raise RuntimeError(f"No videos found in {self.root}. Check split & extensions.")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        vid_path, label = self.samples[idx]
        frames = sample_frames_uniform(str(vid_path), num_frames=T)
        frames = [img_tf(cv2.cvtColor(f, cv2.COLOR_BGR2RGB)) for f in frames]
        x = torch.stack(frames, dim=0)  # (T, C, H, W)
        y = torch.tensor(label, dtype=torch.long)
        return x, y

BATCH_SIZE = 8
train_ds = VideoFolderDataset("/content/dataset", "train")
val_ds   = VideoFolderDataset("/content/dataset", "val")
test_ds  = VideoFolderDataset("/content/dataset", "test")

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

print("Train/Val/Test sizes:", len(train_ds), len(val_ds), len(test_ds))
x, y = next(iter(train_loader))
print("Batch x:", x.shape, "y:", y.shape)
```

```python
import torch.nn as nn
from torchvision import models

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", DEVICE)

class CNNLSTM(nn.Module):
    def __init__(self, num_classes=2, lstm_hidden=256, lstm_layers=1, dropout=0.3):
        super().__init__()
        weights = models.ResNet50_Weights.IMAGENET1K_V2
        backbone = models.resnet50(weights=weights)
        # remove classifier: keep everything except the final fc
        self.cnn = nn.Sequential(*(list(backbone.children())[:-1]))  # output: (B, 2048, 1, 1)
        self.feat_dim = 2048

        self.lstm = nn.LSTM(
            input_size=self.feat_dim,
            hidden_size=lstm_hidden,
            num_layers=lstm_layers,
            batch_first=True,
            bidirectional=False,
            dropout=0.0 if lstm_layers == 1 else dropout,
        )
        self.head = nn.Sequential(
            nn.Dropout(dropout),
            nn.Linear(lstm_hidden, num_classes)
        )

    def forward(self, x):
        # x: (B, T, C, H, W)
        B, T, C, H, W = x.shape
        x = x.view(B*T, C, H, W)
        feats = self.cnn(x).view(B*T, self.feat_dim)     # (B*T, 2048)
        feats = feats.view(B, T, self.feat_dim)          # (B, T, 2048)
        out, _ = self.lstm(feats)                        # (B, T, hidden)
        last = out[:, -1, :]                             # (B, hidden)
        logits = self.head(last)                         # (B, num_classes)
        return logits

model = CNNLSTM().to(DEVICE)

# quick forward
xb, yb = next(iter(train_loader))
with torch.no_grad():
    logits = model(xb.to(DEVICE))
print("Logits:", logits.shape)
```

```python
from tqdm import tqdm

def accuracy_from_logits(logits, y):
    preds = logits.argmax(dim=1)
    return (preds == y).float().mean().item()

def run_epoch(model, loader, optimizer=None):
    train = optimizer is not None
    model.train() if train else model.eval()

    total_loss, total_acc, n = 0.0, 0.0, 0
    criterion = nn.CrossEntropyLoss()

    for x, y in tqdm(loader, leave=False):
        x, y = x.to(DEVICE, non_blocking=True), y.to(DEVICE, non_blocking=True)

        if train:
            optimizer.zero_grad(set_to_none=True)

        with torch.set_grad_enabled(train):
            logits = model(x)
            loss = criterion(logits, y)
            if train:
                loss.backward()
                optimizer.step()

        bs = x.size(0)
        total_loss += loss.item() * bs
        total_acc  += accuracy_from_logits(logits, y) * bs
        n += bs

    return total_loss / n, total_acc / n
```

```python
# Freeze CNN
for p in model.cnn.parameters():
    p.requires_grad = False

trainable = [p for p in model.parameters() if p.requires_grad]
optimizer = torch.optim.Adam(trainable, lr=1e-3)

EPOCHS_HEAD = 5
for epoch in range(EPOCHS_HEAD):
    tr_loss, tr_acc = run_epoch(model, train_loader, optimizer=optimizer)
    va_loss, va_acc = run_epoch(model, val_loader, optimizer=None)
    print(f"[HEAD] epoch {epoch+1}/{EPOCHS_HEAD} | train acc={tr_acc:.3f} val acc={va_acc:.3f}")
```

```python
# Unfreeze last part of CNN (roughly the last 1/4 of parameters)
params = list(model.cnn.parameters())
for p in params[int(len(params)*0.75):]:
    p.requires_grad = True

trainable = [p for p in model.parameters() if p.requires_grad]
optimizer = torch.optim.Adam(trainable, lr=1e-5)

EPOCHS_FT = 10
for epoch in range(EPOCHS_FT):
    tr_loss, tr_acc = run_epoch(model, train_loader, optimizer=optimizer)
    va_loss, va_acc = run_epoch(model, val_loader, optimizer=None)
    print(f"[FT]   epoch {epoch+1}/{EPOCHS_FT} | train acc={tr_acc:.3f} val acc={va_acc:.3f}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix

model.eval()
all_preds, all_labels = [], []
with torch.no_grad():
    for x, y in tqdm(test_loader, leave=False):
        logits = model(x.to(DEVICE))
        preds = logits.argmax(dim=1).cpu().numpy().tolist()
        all_preds.extend(preds)
        all_labels.extend(y.numpy().tolist())

print("Classification report:")
print(classification_report(all_labels, all_preds, target_names=["correct","wrong"]))
print("Confusion matrix:")
print(confusion_matrix(all_labels, all_preds))
```
